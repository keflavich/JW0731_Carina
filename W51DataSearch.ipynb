```python
from astroquery.ipac.irsa import Irsa
```

```python
from astropy import units as u, coordinates
```

```python
coords = coordinates.SkyCoord('19:23:43.1888 14:30:50.004', frame='icrs', unit=(u.h, u.deg))
```

```python
def try_get_cat(cat):
    try:
        return Irsa.query_region(coords, catalog=cat, radius=3*u.arcsec)
    except:
        return
```

```python
cats = {cat:
        try_get_cat(cat)
        for cat in Irsa.list_catalogs()
       }
```
